```python
RFMScores = dataset.groupby('CustomerID').agg({'Date' : lambda x: (latest_date - x.max()).days,
                                               'Invoice' : lambda x: x.count(),
                                               'Price' : lambda x: x.sum()})
```

```python
# Split the data into the quantile methods
quantiles = RFMScores.quantile(q=[0.25,0.50,0.75])
quantiles = quantiles.to_dict()
quantiles
```

```python
def Recencyscore (x,p,d):
    if x<=d[p][0.25]:
        return 1
    elif x<=d[p][0.50]:
        return 2
    elif x<=d[p][0.75]:
        return 3
    else :
        return 4
```

```python
def Frequencyscore (x,p,d):
    if x<=d[p][0.25]:
        return 4
    elif x<=d[p][0.50]:
        return 3
    elif x<=d[p][0.75]:
        return 2
    else :
        return 1
```

```python
def Monetoryscore (x,p,d):
    if x<=d[p][0.25]:
        return 4
    elif x<=d[p][0.50]:
        return 3
    elif x<=d[p][0.75]:
        return 2
    else :
        return 1
```

```python
RFMScores.columns
```

```python
RFMScores['R'] = RFMScores['Recency'].apply(Recencyscore,args = ('Recency', quantiles))
RFMScores['F'] = RFMScores['Frequency'].apply(Frequencyscore,args = ('Frequency', quantiles))
RFMScores['M'] = RFMScores['Monetory'].apply(Monetoryscore,args = ('Monetory', quantiles))
```

```python
RFMScores.reset_index()
```

```python
RFMScores['RFMvalue'] = RFMScores[['R','F','M']].sum(axis = 1)
RFMScores['RFMGroup'] = RFMScores.R.map(str) + RFMScores.F.map(str) + RFMScores.M.map(str)
RFMScores.reset_index()
```

```python
RFMScores['RFMvalue'].nunique()
```

```python
Loyalty_lavel = ['Platinum','Diamond','Gold','Silver']
Scores_cuts = pd.qcut(RFMScores.RFMvalue, q=4 ,labels =Loyalty_lavel)
RFMScores['Loyalty_lavel'] = Scores_cuts.values
RFMScores = RFMScores.reset_index()
RFMScores
```

```python
RFMScores.to_csv('Final_data.csv')
```
